```python
import billboard as bb
import pandas as pd
```

```python
dates = list()
songs = list()
artists = list()
peaks = list()
weeks = list()
```

```python
chart = bb.ChartData('hot-100')
```

```python
while chart.date >= '2006-01-01':

    if int(chart.date[8:10]) <= 7:
        
        for i in range(0,100):
            dates.append(chart.date)
            songs.append(chart[i].title)
            artists.append(chart[i].artist)
            peaks.append(chart[i].peakPos)
            weeks.append(chart[i].weeks)
        
    chart = bb.ChartData('hot-100', chart.previousDate)
```

```python
all_charts = pd.DataFrame(
{
    'date': dates,
    'song': songs,
    'artist': artists,
    'peakPos': peaks,
    'weeks': weeks
})
```
